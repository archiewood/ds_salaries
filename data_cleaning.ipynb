```python
import pandas as pd
```

```python
df=pd.read_csv('salary_data.csv')
```

```python
#drop rows where 'title' column does not contain the string 'title'
df=df[df['title'].str.contains('title', case=False)]

#remove duplicate rows that match all columns except fpr 'page'
df=df.drop_duplicates(subset=['title','salary','tenure','location'], keep='first')
```

```python
df.head(15)
```

```python
#remove "•" character from all columns
df=df.replace(to_replace='•', value='', regex=True)
#trim whitespace from all string columns
df=df.apply(lambda x: x.str.strip() if x.dtype == 'object' else x)
#make all columns lowercase, and remove text inside parentheses
df=df.apply(lambda x: x.str.lower().str.replace('\(.*\)', '') if x.dtype == 'object' else x)
```

```python
df
```

```python
# Remove "Title: " from title column
df['title']=df['title'].str.replace('Title:', '', case=False)
# Remove "Salary: " from salary column
df['salary']=df['salary'].str.replace('Salary:', '', case=False)
# Remove "Location: " from location column
df['location']=df['location'].str.replace('Location:', '', case=False)
# Remove "Tenure length: " from tenure column
df['tenure']=df['tenure'].str.replace('Tenure length:', '', case=False)
# Remove "Company/Industry: " from industry column
df['industry']=df['industry'].str.replace('Company/Industry:', '', case=False)
df['industry']=df['industry'].str.replace('Industry:', '', case=False)
# Remove "Education: " from education column
df['education']=df['education'].str.replace('Education:', '', case=False)
# Remove "Prior experience: " from prior_experience column
df['prior_experience']=df['prior_experience'].str.replace('Prior experience:', '', case=False)
# Remove "Total comp: " from total_comp column
df['total_comp']=df['total_comp'].str.replace('Total comp:', '', case=False)
# Remove "Stock and/or recurring bonuses: " from stock_or_bonus column
df['stock_or_bonus']=df['stock_or_bonus'].str.replace('Stock and/or recurring bonuses:', '', case=False)
# Remove "Relocation/Signing Bonus:" from signing_bonus column
df['signing_bonus']=df['signing_bonus'].str.replace('Relocation/Signing Bonus:', '', case=False)

# replace character U+2060 with space in all string columns
df=df.replace(to_replace='\u2060', value='', regex=True)

#trim whitespace from all string columns
df=df.apply(lambda x: x.str.strip() if x.dtype == 'object' else x)

```

```python
# clean titles

# remove "." "," "-" "/", "*" from titles
df.title = df.title.str.replace(r"\.", "").str.replace(
    r",", "").str.replace(r"-", "").str.replace(r"/", " ").str.replace(r"\*", "")

# remove "title :" from titles 
df.title = df.title.str.replace(r"title :", "")

#replace mle with machine learning engineer in titles
df.title = df.title.str.replace(r"mle", "machine learning engineer")

#replace ml with machine learning in titles
df.title = df.title.str.replace(r"ml", "machine learning")

#replace ds with data scientist in titles
df.title = df.title.str.replace(r" ds", " data scientist")

# replace "sr" with "senior" and "jr" with "junior" in titles
df.title = df.title.str.replace(
    r"sr ", "senior ").str.replace(r"jr ", "junior ")

# replace numbers with lowercase romans in titles
df.title = df.title.str.replace(r"1", "i").str.replace(r"2", "ii").str.replace(r"3", "iii").str.replace(r"4", "iv")

#strip whitespace from titles
df.title = df.title.str.strip()

# replace multiple spaces with single space in titles
df.title = df.title.str.replace(r"\s+", " ")
```

```python
# clean tenure

# remove "tenure:" and "yoe:" from tenure column
df.tenure = df.tenure.str.replace(r"tenure:", "").str.replace(r"yoe:", "")


# remove non numeric characters except "."" or "," from tenure, convert to numeric
df['tenure_clean'] = df.tenure.str[0:7].str.replace(r"[^0-9\.,]", "").str.replace(r", ", "")


# try to convert to numeric
df['tenure_clean'] = pd.to_numeric(df['tenure_clean'], errors='coerce')
# if df.tenure contains "month" then divide df.tenure_clean by 12
df['tenure_clean'] = df.tenure_clean.where(df.tenure.str.contains("mo") == False, df.tenure_clean/12.0, errors='ignore')
# if tenure_clean > 40 then replace with Null
df['tenure_clean'] = df.tenure_clean.where(df.tenure_clean <= 40, None)
```

```python
# clean salary data

# remove the word "base", "salary" and "tax" from salary
df['salary_pre_clean'] = df.salary.str.replace(r"(?i)base", "").str.replace(r"(?i)salary", "").str.replace(r"(?i)tax", "").str.replace(r"(?i)currency", "").str.replace(r"(?i)current", "")

# replace "Lakhs" with "L" in salary_pre_clean\
df['salary_pre_clean'] = df.salary_pre_clean.str.replace(r"(?i)lakhs", "L")
# replace "lakh" with "L" in salary_pre_clean
df['salary_pre_clean'] = df.salary_pre_clean.str.replace(r"(?i)lakh", "L")
df['salary_pre_clean'] = df.salary_pre_clean.str.replace(r"(?i)lpa", "L")


# remove any characters after ", "
df['salary_pre_clean'] = df.salary_pre_clean.str.replace(r", .*", "")


# remove non numeric characters except "." from salary, convert to numeric
df['salary_clean'] = pd.to_numeric(df.salary_pre_clean.str[0:12].str.replace(r"[^0-9\.]", ""), errors='coerce')




# multiply by 1000 if contains k
df['salary_clean'] = df.salary_clean.where(
    df.salary_pre_clean.str.contains("k") == False, df.salary_clean*1000, errors='ignore').where(
        df.salary_pre_clean.str.contains("K ") == False, df.salary_clean*1000, errors='ignore').where(
                df.salary_pre_clean.str.contains("K$") == False, df.salary_clean*1000, errors='ignore').where(
                    df.salary_pre_clean.str.contains("mon") == False, df.salary_clean*12, errors='ignore').where(
                            df.salary_pre_clean.str.contains("/mo") == False, df.salary_clean*12, errors='ignore')



df['salary_clean'] = df.salary_clean.where(df.salary_pre_clean.str.contains("L") == False, df.salary_clean*100000, errors='ignore')
df['salary_clean'] = df.salary_clean.where(df.salary_pre_clean.str.contains("dkk") == False, df.salary_clean/1000, errors='ignore')
```
